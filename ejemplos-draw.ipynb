# Ejemplos del método Draw() de YAGS en libretas Jupiter

```gap
LoadPackage("yags");
```

```gap
C:=CompleteGraph(5);
```

 Configuramos las coordenadas de la gráfica:

```gap
SetCoordinates(C,[ [ -237, -28 ], [ -27, -242 ], [ 238, -105 ], [ 193, 190 ], [ -102, 238 ] ]);
```

### Primera versión de Draw 

 Esta es la versión del comando Draw de 1 sólo argumento. **Para poder habilitar los comandos tradicionales**, basta con dar click en cualquier parte del área del recuadro donde se está dibujando la gráfica (esto también ilumina con naranja, la línea del recuadro). Los comandos se desactivan al dar click en cualquier otra parte fuera del recuadro de la página. 

```gap
Draw(C);
```

### Segunda versión de Draw

Esta es la segunda versión del comando Draw (disponible desde la versión 0.0.4 de YAGS). El segundo argumento es una arreglo de vértices que se van a iluminar de un color diferente de los demás vértices (por default, el color para resultarlos es rojo).

```gap
Draw(C,[1,2,3]);
```

### Propuesta para una tercera versión del comando Draw

Esta es una propuesta para una **tercera versión** del comando Draw. El método es de la forma:

<p style="text-align: center;">Draw(gráfica, <strong>coloracionVertices</strong>)</p>


Donde **coloracionVertices** es un *record* en el que *cada propiedad* es un **código de un color en hexadecimal**
y el valor asociado a dicha *propiedad* son los vertices que se van a iluminar de dicho color.

Por ejemplo, para iluminar a los vértices **[1,2,3]** de la gráfica **C** con el color **16871b** (una variación del verde), se usa el siguiente comando:

```gap
Draw(C,rec(16871b:=[1,2,3]));
```

Si además queremos que los vértices **[4,5]** sean de color **FFA500** (una variación del naranja), invocamos a Draw de la siguiente forma:

```gap
Draw(C,rec(16871b:=[1,2,3],FFA500:=[4,5]));
```

### Propuesta para una cuarta versión del comando Draw

Esta es una propuesta para una **cuarta versión** del comando Draw. El método es de la forma:

<p style="text-align: center;">
Draw(gráfica, <strong>coloracionVertices</strong>, <strong>coloracionAristas</strong>)
</p>


Donde **coloracionVertices** es un *record* que se define igual que en la **versión 3** y **coloracionAristas** es un record en el que cada *propiedad* es el código de un color en hexadecimal y su valor asociado es un arreglo de las aristas a iluminar.

Por ejemplo, si queremos iluminar los *vértices con los mismos colores* que se usaron en el ejemplo de la **versión 3** y además, iluminar las aristas *\[ \[1,2\],\[2,3\],\[3,1\] \]* con el color **006400** (una variación del verde)  y la arista *\[4,5\]* con el color **f5ed00** (una variación del amarillo), invocamos a Draw de la siguiente forma:

```gap
Draw(C,rec(16871b:=[1,2,3],FFA500:=[4,5]),rec(006400:=[[1,2],[2,3],[3,1]],f5ed00:=[[4,5]]));
```
